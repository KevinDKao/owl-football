```python
import sys
import os

# Add the 'data' folder to the Python path
import sys

sys.path.append("..")
from data.games import get_games

import pandas as pd
import time  # Import the time module for adding delays

# Define lists for session_type and classification
years = list(range(2000, 2026))  # From 2000 to 2025
session_types = [
    "regular",
    "postseason",
    "both",
    "allstar",
    "spring_regular",
    "spring_postseason",
]
classifications = ["fbs", "fcs", "ii", "iii"]

# Initialize an empty DataFrame to store the results
games_df = pd.DataFrame()

# Iterate over all years
for year in years:
    # Initialize a list to store the data for the current year
    year_games_data = []

    # Print the message for the current year
    print(f"Fetching data for Year: {year}")

    # Iterate over all session types and classifications
    for session_type in session_types:
        for classification in classifications:
            # Make the API call with the current parameters (remove 'week')
            games_data = get_games(
                year=year, session_type=session_type, classification=classification
            )

            # If games_data is not empty, add it to the year-specific list
            if games_data:
                year_games_data.extend(games_data)

    # If there is data for the year, convert it to a DataFrame and append it to the main DataFrame
    if year_games_data:
        year_df = pd.DataFrame(year_games_data)
        games_df = pd.concat([games_df, year_df], ignore_index=True)

    # Add a delay after processing each year
    print(f"Finished fetching data for Year {year}. Pausing for 15 seconds...\n")
    time.sleep(15)

# Display the first few rows of the combined data
print("\nGames Data:")
print(games_df.head())
```
